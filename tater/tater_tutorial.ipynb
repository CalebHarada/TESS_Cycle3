```python
import tater
```

```python
# choose a planet and initialize
#tic_id = 230127302 # TOI-1246
tic_id = 43647325 # WASP-35 b
transit_fitter = tater.TransitFitter(tic_id, auto_params=True) # automatically find stellar params on MAST
```

```python
# download data, show plot
transit_fitter.download_data(n_sectors=1, show_plot=True)
```

```python
# find planets
planets = transit_fitter.find_planets(show_plots=True)
```

```python
# MCMC fit
transit_fitter.fit_transits()
```
